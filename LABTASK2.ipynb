## Lab Task: Data Visualization with Matplotlib

# Task 1: Basic Line Plot

1.1.1 Create a basic line plot for the following data points
o x = [1, 2, 3, 4, 5]
o y = [2, 4, 6, 8, 10]
Add labels for both axes, title, and grid lines

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.plot(x, y, marker='o')


plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')


plt.title('Basic Line Plot')

plt.grid()


plt.show()
```

# Task 2: Scatter Plot

1.2.1 Create a scatter plot using the following data points:
o x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
o y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
Customize the markers with size and color

```python
import matplotlib.pyplot as plt


x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]


plt.scatter(x, y, s=100, color='blue', alpha=0.7, edgecolor='black')  # Customize size and color


plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Add a title
plt.title('Customized Scatter Plot')


plt.grid()
plt.show()
```

# Task 3: Bar Plot

1.3.1 Create a bar plot with the following categories and values:
o categories = ['A', 'B', 'C', 'D', 'E']
o values = [5, 7, 3, 8, 6]
Customize the bars with different colors for each category.

```python
import matplotlib.pyplot as plt


categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]


colors = ['red', 'blue', 'green', 'orange', 'purple']


plt.bar(categories, values, color=colors)


plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Customized Colors')


plt.grid(axis='y')
plt.show()
```

Task 4: Histogram
# New Section

1.4.1 Generate a histogram for a dataset containing random numbers:
o Use np.random.randn(1000) to generate random numbers.
o Create a histogram with 30 bins and customize the color of the bars.

```python
import numpy as np
import matplotlib.pyplot as plt


data = np.random.randn(1000)


plt.hist(data, bins=30, color='skyblue', edgecolor='black')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')


plt.grid(axis='y', alpha=0.75)


plt.show()
```

# Task 5: Pie Chart

1.5.1 Create a pie chart for the following data:
o labels = ['Python', 'Java', 'C++', 'JavaScript']
o sizes = [50, 25, 15, 10]
Add percentages to each section and display the chart with an exploded view for the
largest section.

```python
import matplotlib.pyplot as plt


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]


explode = (0.1, 0, 0, 0)


plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])


plt.title('Programming Language Popularity')


plt.axis('equal')
plt.show()
```
